```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('delhivery.csv')
df.head()
```

```python
missing_values = df.isnull().sum()
print("Missing values per columns:\n" , missing_values)
```

```python
# No missing values in numeric columns in this dataset
```

```python
#handling missing values in Categorial Columns 
df['source_name'].fillna('missing', inplace=True)
df['destination_name'].fillna('missing',inplace = True)
print(df[['source_name', 'destination_name']].head(20))
```

```python
print(df.dtypes)
```

```python
#CONVERSION OF DATE-RELATED COLUMNS TO DATETIME FORMAT
datetime_columns = ['trip_creation_time', 'od_start_time', 'od_end_time', 'cutoff_timestamp']
for col in datetime_columns:
    df[col] = pd.to_datetime(df[col],errors='coerce')

df.dropna(subset=['od_start_time','od_end_time'], inplace = True)
    
#CONVERSION OF CATEGORIAL COLUMNS TO CATEGORY TYPE
categorical_columns = ['data','route_type','source_center','destination_center','source_name','destination_name']
for col in categorical_columns:
    df[col] = df[col].astype('category')

print(df.dtypes)
# Display the first 20 rows of the updated dataset
df.head(20)
```

```python
#calculate Trip Duration in minutes 
df['trip_duration'] = (df['od_end_time'] - df['od_start_time']).dt.total_seconds() / 60
df['distance_covered'] = df['actual_distance_to_destination']

#extract day of the week and month from the trip creation time
df['day_of_week'] = df['trip_creation_time'].dt.day_name()
df['month'] = df['trip_creation_time'].dt.month_name()

df.head(30)
```

```python
nan_rows = df[df[['trip_creation_time', 'od_start_time', 'od_end_time']].isnull().any(axis=1)]
print("Rows with NaN in date-related columns:\n", nan_rows)

# Replace NaN values in trip_duration with the mean
df['trip_duration'].fillna(df['trip_duration'].mean(), inplace=True)

#Replace NaN values in day_of_week with the mode 
df['day_of_week'].fillna(df['day_of_week'].mode()[0],inplace=True)

#Replace NaN values in month with the mode 
df['month'].fillna(df['month'].mode()[0], inplace = True)

df.head(10)
```

```python
df.info()
```

```python
# calculate Q1 and Q3 for trip_duration

Q1 = df['trip_duration'].quantile(0.25)
Q3 = df['trip_duration'].quantile(0.75)
IQR = Q3 - Q1


#Define the lower and upper bounds 
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#indentify outrliers in trip_duration
df['is_trip_duration_outlier'] = (df['trip_duration'] < lower_bound ) | (df['trip_duration'] > upper_bound)

# filter out outliers 
df = df[~df['is_trip_duration_outlier']]
```

```python
print("Remaining outliers:", df['is_trip_duration_outlier'].sum())
```

```python
df.head()
```

```python
# print(df.columns)
# Standardize column names to follow a consistent format 
df.columns = df.columns.str.lower().str.replace(' ', '_')
print(df.columns)
```

```python
# Normalize Numeric Columns with Min-Max Scaling:
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

numeric_cols = df.select_dtypes(include=['float64','int64']).columns

#apply MinMax scaling to numeric columns and update the  dataframe
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

#display the scaled data
df.head()
```
